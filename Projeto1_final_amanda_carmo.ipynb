## Explicação da hipótese e mecanismo

O grupo é representado por países emergentes, ou seja, cujas economias naconjuntura atual são semelhantes, possuindo uma potencialidade economica elevada, se comparados aos demais países, também emergentes, em suas regiões.
   <br>   
Após a formação dos BRICS, os países começaram a apoiar-se mutuamente, segundo a 61º Assembleia Geral das Nações Unidas, direcionada também ao melhoramento da comunicação entre os países emergentes, além de buscar forte cooperação. Desta forma, é possível enxergar um crescimento e estabilização concomitantes.

   <br>
   Tal como informado no site oficial do grupo, os BRICS foram formads em 2006, por países emergentes relevantes economicamente, participantesde acordos e organizações internacionais, tais como a ONU e o G20. Desde sua formação, como relatado no portal do Itamaraty, a partir da realização em 2009 da primeira cúpula, o BRICS adquiriu notável expansão em suas atividades, direcionadas a diversos âmbitos. Todavia, nota-se principalmente a expansão no âmbito financeiro, fato este que motivou o lançamento da instituição "O Novo Banco de Desenvolvimento", cujo objetvo foi lidar com a falta de recursos financeiros voltados a projetos de infraestrutura.

## Importação (% PIB)

### Cruzamento de dados;

Foram cruzados os dados referentes à exportação, no eixo das ordenadas e a taxa de emprego no eixo das abscissas. Visa-se checar se há uma correlação grande entre tais variáveis.

```python
def reta(dados_x, dados_y, c):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()

    print('Ajuste linear entre X3 e X4:')
    print('Coeficiente linear:', b)
    print('Coeficiente angular:', a)
    plt.scatter(dados_x, dados_y, c=c, alpha=0.8)
    plt.title('')
    plt.xlabel('taxa de emprego')
    plt.ylabel('exportação')

    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color=c)

    plt.tight_layout()
    
dados_x = emprego
dados_y = exports
```
